```python
import imageio
import numpy as np

import matplotlib.pyplot as plt

image = imageio.imread("results/actual_binary.png")
```

```python
plt.imshow(image)
```

```python
def highest_pixels(img, threshold=0):
    highest = []
    for i in range(img.shape[1]):
        column = img[:, i, 0]
        wanted = np.where(column > threshold)[0]
        if len(wanted) == 0:
            highest.append(0)
        else:
            highest.append(wanted[0])
            
    return highest
```

```python
q = highest_pixels(image)
```
